```r
library(ComplexHeatmap)
library(tidyverse)
library(circlize)
```

### Mild

```r
cor_met_asv = rcorr(as.matrix(mild_asv), as.matrix(mild_met), type = "pearson")
r = cor_met_asv$r
p = cor_met_asv$P
pmt = as.matrix(p)

write.csv(pmt, "相关性热图/202506热图/mild_asv_met_p.csv", row.names = TRUE)
write.csv(r, "相关性热图/202506热图/mild_asv_met_r.csv", row.names = TRUE)

cor_met_asv = rcorr(as.matrix(mil_asv), as.matrix(mil_gene), type = "pearson")
r = cor_met_asv$r
p = cor_met_asv$P
pmt = as.matrix(p)

write.csv(pmt, "相关性热图/202506热图/mild_asv_gene_p.csv", row.names = TRUE)
write.csv(r, "相关性热图/202506热图/mild_asv_gene_r.csv", row.names = TRUE)
```

```r
mil_asv_met_r = read.csv("相关性热图/202506热图/mild_asv_met_r.csv", row.names = 1)
mil_asv_met_p = read.csv("相关性热图/202506热图/mild_asv_met_p.csv", row.names = 1)

pmt = as.matrix(mil_asv_met_p)

if(!is.null(pmt)){
    sssmt = pmt < 0.001
    pmt[sssmt] = "***"
    ssmt = pmt < 0.01 & pmt >= 0.001
    pmt[ssmt] = "**"
    smt = pmt < 0.05 & pmt >= 0.01
    pmt[smt] = "*"
    pmt[!sssmt&!ssmt&!smt] = ""
} else {
   pmt = F
}
#pdf("相关性热图/202506热图/heatmap_mild_asv_met.pdf", width=7, height=7)
colors = colorRamp2(c(-1,0,1),c("#54afd9","white","#f69aa3"))
Heatmap(as.matrix(mil_asv_met_r), name = "R", show_row_names = T, 
        show_column_names = T, cluster_rows = F, 
        cluster_columns = F,
        # row_split = r_group$group,
        layer_fun = function(j, i, x, y, width, height, fill){
                grid.text(sprintf("%s", pindex(pmt, i, j)), 
                          x, 
                          y, 
                          gp = gpar(fontsize = 20))
        },
        row_names_side = "left",col = colors)
#dev.off()
```

```r
mil_asv_gene_r = read.csv("相关性热图/202506热图/mild_asv_gene_r.csv", row.names = 1)
mil_asv_gene_p = read.csv("相关性热图/202506热图/mild_asv_gene_p.csv", row.names = 1)

pmt = as.matrix(mil_asv_gene_p)

if(!is.null(pmt)){
    sssmt = pmt < 0.001
    pmt[sssmt] = "***"
    ssmt = pmt < 0.01 & pmt >= 0.001
    pmt[ssmt] = "**"
    smt = pmt < 0.05 & pmt >= 0.01
    pmt[smt] = "*"
    pmt[!sssmt&!ssmt&!smt] = ""
} else {
   pmt = F
}
#pdf("相关性热图/202506热图/heatmap_mild_asv_gene.pdf", width=7, height=7)
colors = colorRamp2(c(-1,0,1),c("#54afd9","white","#f69aa3"))
Heatmap(as.matrix(mil_asv_gene_r), name = "R", show_row_names = T, 
        show_column_names = T, cluster_rows = F, 
        cluster_columns = F,
        # row_split = r_group$group,
        layer_fun = function(j, i, x, y, width, height, fill){
                grid.text(sprintf("%s", pindex(pmt, i, j)), 
                          x, 
                          y, 
                          gp = gpar(fontsize = 20))
        },
        row_names_side = "left",col = colors)
#dev.off()
```
